```python
users
```

```python
# 2. What is the result of using a right join on the DataFrames?

pd.merge(users,roles, 
         how='right', 
         left_on='role_id', 
         right_on='id')
```

```python
# 3. What is the result of using an outer join
# on the DataFrames?

pd.merge(users,roles, how='outer', left_on='role_id', right_on='id',indicator = True)

```

```python
# 12. Create a column named mileage_difference like you 
# did in the DataFrames exercises; this column should 
# contain the difference between highway and city mileage 
# for each car.

mpg['milage_difference'] = (mpg.highway - mpg.city).abs()

mpg
```

```python
# 13. Create a column named average_mileage like you did 
# in the DataFrames exercises; this is the mean of the city
# and highway mileage.

mpg['average_milage'] = mpg[['city', 'highway']].mean(axis=1)

mpg
```

```python
# 15. Using the mpg dataset, find out which which 
# manufacturer has the best miles per gallon on average?

best_mpg = mpg.groupby(['manufacturer'])['average_milage'].mean()

best_mpg
```
